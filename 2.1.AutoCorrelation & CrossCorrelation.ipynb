```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate as cr 
```

```python
x=[5,4,7,6,9,2]
y=[2,8,2,4,9,3]
z=cr(x,y)
print(z)
```

```python
t=np.arange(0,len(z))
print(t)
```

```python
plt.plot(t,z)
plt.show()
```

```python
t=np.linspace(0,1,1000)
s1=np.sin(2*np.pi*1000*t)
s2=np.sin((2*np.pi*1000*t)+np.pi/4)
plt.plot(t,s1)
plt.plot(t,s2)
plt.show()
```

```python
z=cr(s1,s2)
print(z)
plt.plot(z)
plt.show()
```

```python
t1=np.linspace(0,10,1000)
t2=np.linspace(0,20,2000)
s1=np.sin(2*np.pi*1000*t1)
s2=np.sin(2*np.pi*1000*t2)
plt.plot(t1,s1)
plt.plot(t2,s2)
```
