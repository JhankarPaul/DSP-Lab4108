```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate as cr 
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
z=cr(x,x)
b=cr(x,y)
#print(z)
```

```python
t=np.arange(0,len(z))
print(t)
```

```python
plt.plot(t,z)
plt.title("auto correlation")
plt.show()
plt.plot(t,b)
plt.title("cross correlation")
plt.show()
```

```python
t=np.linspace(0,1,1000)
s1=np.sin(2*np.pi*1000*t)
s2=np.sin((2*np.pi*1000*t)+np.pi/4)
plt.plot(t,s1)
plt.plot(t,s2)
plt.show()
```

```python
z=cr(s1,s2)
print(z)
plt.plot(z)
plt.show()
```

```python
f=1000
t1=np.linspace(0,1,1000)
s1=np.sin(2*np.pi*f*t1)
plt.subplot(2,3,1)
plt.plot(t1,s1)

t2=np.linspace(0.5,1.5,1000)
s2=np.sin(2*np.pi*f*t2)
plt.subplot(2,3,2)
plt.plot(t2,s2)

t3=np.linspace(0.5,1.5,1000)
s3=np.cos(2*np.pi*f*t3)
plt.subplot(2,3,3)
plt.plot(t3,s3)

s4=cr(s1,s2,mode='same')
plt.subplot(2,3,4)
plt.plot(s4)
plt.title("auto")

s5=cr(s1,s3,mode='same')
plt.subplot(2,3,5)
plt.plot(s5)
plt.title("cross")
```
