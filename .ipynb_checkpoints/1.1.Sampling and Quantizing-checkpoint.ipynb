```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 1, 0.0001)
y = 5 * np.sin(2 * np.pi * 5 * t)

plt.subplot(6, 1, 1)
plt.plot(t, y)
plt.title('Main Signal')

fs = 5
t1 = np.arange(0, 1, 1 / fs)
y = 5 * np.sin(2 * np.pi * 5 * t1)
plt.subplot(6, 1, 2)
plt.plot(t1, y)
plt.title('Sampling Frequency= 5Hz')

fs = 10
t2 = np.arange(0, 1, 1 / fs)
y = 5 * np.sin(2 * np.pi * 5 * t2)
plt.subplot(6, 1, 3)
plt.plot(t2, y)
plt.title('Sampling Frequency= 10Hz')

fs = 100
t3 = np.arange(0, 1, 1 / fs)
y = 5 * np.sin(2 * np.pi * 5 * t3)
plt.subplot(6, 1, 4)
plt.plot(t3, y)
plt.title('Sampling Frequency= 100Hz')

fs = 1000
t4 = np.arange(0, 1, 1 / fs)
y = 5 * np.sin(2 * np.pi * 5 * t4)
plt.subplot(6, 1, 5)
plt.plot(t4, y)
plt.title('Sampling Frequency= 1000Hz')

B = np.zeros(len(y))

for i in range(len(y)):
    B[i] = y[i]
    if B[i] <= -3.75:
        B[i] = -3.75
    elif -3.75 < B[i] <= -2.50:
        B[i] = -2.50
    elif -2.50 < B[i] <= -1.25:
        B[i] = -1.25
    elif -1.25 < B[i] <= 0.00:
        B[i] = 0.00
    elif 0.00 < B[i] <= 1.25:
        B[i] = 1.25
    elif 1.25 < B[i] <= 2.50:
        B[i] = 2.50
    elif 2.50 < B[i] <= 3.75:
        B[i] = 3.75
    elif 3.75 < B[i] <= 5.00:
        B[i] = 5.00

plt.figure()
plt.subplot(2, 1, 1)
plt.plot(t4, B)
plt.title('8 Bit Sampling')
```
