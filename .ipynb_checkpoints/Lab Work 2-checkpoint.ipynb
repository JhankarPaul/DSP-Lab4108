```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import correlate as cr 
```

```python
x=[1,3,5,7]
y=[8,16,10,11]

acr=cr(x,x)
t=np.arange(len(acr))
plt.plot(t,acr)
plt.title('Auto')
plt.show()

ccr=cr(x,y)
t=np.arange(len(ccr))
plt.plot(t,ccr)
plt.title('Cross')
plt.show()
```

```python
t=np.linspace(0,1,1000)

s1=np.sin(2*np.pi*2*t)
plt.plot(t,s1)
#plt.title("S1 = Sin(4*pi*t)")

s2=np.sin((2*np.pi*2*t)+(np.pi/2))
plt.plot(t,s2)
plt.title("S1 = Sin(4*pi*t) & S2 = Sin(4*pi*t) + pi/2")
plt.show()

s1=np.sin(2*np.pi*2*t)
plt.plot(t,s1)
#plt.title("S1 = Sin(4*pi*t)")

s3=np.cos((2*np.pi*2*t)+(np.pi/4))
plt.plot(t,s3)
plt.title("S1 = Sin(4*pi*t) & S3 = cos(4*pi*t) + pi/4")
plt.show()

acr=cr(s1,s1,mode='same')
plt.plot(t,acr)
plt.title("Auto Correlation between S1 and S1")
plt.show()

acr1=cr(s1,s2,mode='same')
plt.plot(t,acr1)
plt.title("Auto Correlation between S1 and S2")
plt.show()

ccr=cr(s1,s3,mode='same')
plt.plot(t,ccr)
plt.title("Cross Correlation between S1 and S3")
plt.show()
```
