```python
import numpy as np
import matplotlib.pyplot as plt

# Set up the time axis
t = np.linspace(0, 2*np.pi, 1000)

# Create two sine waves with a phase difference of 45 degrees
sine1 = 3*np.sin(2*t)
sine2 = 2*np.sin(2*t + np.pi/4)

# Plot the sine waves
plt.plot(t, sine1, label='Sine 1')
plt.plot(t, sine2, label='Sine 2')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Two Sine Waves with 45 Degree Phase Difference')
plt.show()

# Calculate the cross-correlation between the two sine waves
corr = np.correlate(sine1, sine2, mode='same')

# Plot the cross-correlation
plt.plot(t, corr)
plt.xlabel('Time')
plt.ylabel('Correlation')
plt.title('Cross-Correlation between Sine 1 and Sine 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the signals
a = np.array([1, 2, 3, 2, 1])
b = np.array([3, 2, 1])

# Compute the cross-correlation
corr = np.correlate(a, b, mode='same')

# Plot the signals and their cross-correlation
fig, axs = plt.subplots(3, 1, figsize=(8, 6))

axs[0].plot(a)
axs[0].set_title('Signal a')

axs[1].plot(b)
axs[1].set_title('Signal b')

axs[2].plot(corr)
axs[2].set_title('Cross-correlation')

plt.tight_layout()
plt.show()
```
