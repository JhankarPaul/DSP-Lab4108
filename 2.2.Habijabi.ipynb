```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,1,1000)
s1=np.sin(2*np.pi*5*t)


s2=np.sin((2*np.pi*5*t)-(np.pi/4))
plt.plot(t,s1)
plt.plot(t,s2)
plt.show()

s3=np.cos(2*np.pi*5*t)
plt.plot(t,s1)
plt.plot(t,s3)
plt.show()

acr=np.correlate(s1,s2,mode='same')
plt.plot(t,acr)
plt.show()

ccr=np.correlate(s1,s3,mode='same')
plt.plot(t,ccr)
#plt.plot(t,s3)
#plt.plot(t,s1)
plt.show()
```

```python
from scipy import signal
```

```python
tr1=signal.sawtooth(2*np.pi*15*t+np.pi/8,0.5)
plt.plot(t,tr1)
plt.show()
```

```python
ccr1=np.correlate(s1,tr1,mode='same')
plt.plot(t,ccr1)

plt.show()
```
